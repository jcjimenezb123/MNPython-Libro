```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
df=px.data.iris()
df.head()
```

```python
df.describe()
```

```python
px.scatter(df,x='sepal_width',y='sepal_length')
```

```python
px.scatter(df,x='sepal_width',y='sepal_length',
           color='species',
           size='petal_length')
```

```python
px.scatter(df,x='sepal_width',y='sepal_length',
           color='species',
           size='petal_length',
           hover_data=['petal_width'])
```

[trendline](https://plotly.com/python/linear-fits/)

```python
px.scatter(df,x='sepal_width',y='sepal_length',
           color='species',trendline='ols') #ols=ordinary least squares
```

```python
px.scatter_3d(df,x='sepal_length',y='petal_length',z='petal_width',
              color='species',size='sepal_width')
```

```python
px.scatter_matrix(df,dimensions=['sepal_width','sepal_length',
                                 'petal_width','petal_length'],
                  color='species')
```

[Diagrama de cajas y bigotes](https://es.wikipedia.org/wiki/Diagrama_de_caja)

```python
px.box(df,y='sepal_width')
```

```python
px.box(df,y='sepal_width',points='all')
```

```python
px.box(df,x='species',y='sepal_width',
       points='all',color='species')
```

```python
px.violin(df,x='species',y='sepal_width',box=True,
          points='all',color='species')
```

```python
px.histogram(df,x='sepal_width')
```

```python
fig=px.histogram(df,x='sepal_width',color='species')
fig.update_layout(barmode='overlay')
fig.update_traces(opacity=0.75)
fig.show()
```

```python
px.scatter(df,x='sepal_width',y='sepal_length',
           color='species',
           size='petal_width',
           marginal_y='box',
           marginal_x='histogram',
           trendline='ols')
```

```python
px.density_heatmap(df,x='sepal_length',y='petal_length',
                   z='petal_width',
                   marginal_x='histogram',
                   marginal_y='box')
```

```python
df2=px.data.gapminder().query('year==2007')
df2.head()
```

```python
px.scatter_geo(df2,locations='iso_alpha',
               color='continent',
               hover_name='country',
               size='pop',
               projection='orthographic')
```

```python
df3=px.data.wind()
df3.head()
```

```python
px.line_polar(df3,r='frequency',theta='direction',
              color='strength',line_close=True)
```

```python
df4=px.data.experiment()
df4.head()
```

```python
px.scatter_ternary(df4,a='experiment_1',
                   b='experiment_2',
                   c='experiment_3',
                   hover_name='group',color='gender')
```

```python
df5=px.data.gapminder()
df5.head()
```

```python
px.scatter(df5,x='gdpPercap',y='lifeExp',
           animation_frame='year',
           animation_group='country',
           size='pop',
           color='continent',
           hover_name='country',
           log_x=True,size_max=55,range_x=[100,100000],
           range_y=[25,90])
```
